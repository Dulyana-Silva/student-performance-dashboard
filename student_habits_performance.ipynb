```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("student_habits_performance.csv")

# Set the plot size
plt.figure(figsize=(12, 8))

# Create the heatmap using only numeric columns
sns.heatmap(df.corr(numeric_only=True),        # Only numeric columns
            annot=True,                        # Show correlation values
            cmap="coolwarm",                   # Color scheme
            fmt=".2f",                         # Format to 2 decimal places
            linewidths=0.5)                    # Space between cells

# Add a title
plt.title("Correlation Heatmap: Numeric Variables", fontsize=14)

# Show the plot and save
plt.savefig("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("student_habits_performance.csv")

# Set style
sns.set(style="whitegrid")
plt.figure(figsize=(15, 12))

# Study Hours vs Exam Score
plt.subplot(2, 2, 1)
sns.regplot(x='study_hours_per_day', y='exam_score', data=df, scatter_kws={'alpha':0.5})
plt.title('Study Hours vs Exam Score')

# Sleep Hours vs Exam Score
plt.subplot(2, 2, 2)
sns.regplot(x='sleep_hours', y='exam_score', data=df, scatter_kws={'alpha':0.5})
plt.title('Sleep Hours vs Exam Score')

# Social Media Hours vs Exam Score
plt.subplot(2, 2, 3)
sns.regplot(x='social_media_hours', y='exam_score', data=df, scatter_kws={'alpha':0.5})
plt.title('Social Media Hours vs Exam Score')

# Mental Health Rating vs Exam Score
plt.subplot(2, 2, 4)
sns.regplot(x='mental_health_rating', y='exam_score', data=df, scatter_kws={'alpha':0.5})
plt.title('Mental Health Rating vs Exam Score')

plt.xlabel("Study Hours Per Day")
plt.ylabel("Exam Score")

plt.tight_layout()  # Neatly fit all labels and title

# Save to file
plt.savefig("regression_lines.png")  #PNG file will be saved in your current folder
plt.show()
```

```python
# Create a scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='study_hours_per_day', y='exam_score', color='purple')

# Add title and axis labels
plt.title("Study Hours vsExam Score", fontsize=14)
plt.xlabel("Study Hours Per Day")
plt.ylabel("Exam Score")

# Visualize the relationship between Sleep Hours and Exam Score 
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sleep_hours', y='exam_score', color='orange')

# Add title and axis labels
plt.title("Sleep Hours vs Exam Score", fontsize=14)
plt.xlabel("Sleep Hours Per Day")
plt.ylabel("Exam Score")

# Visualize the relationship between Social Media Hours and Exam Score 
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='social_media_hours', y='exam_score', color='blue')

# Add title and axis labels
plt.title("Social Media Hours vs Exam Score", fontsize=14)
plt.xlabel("Social Media Hours Per Day")
plt.ylabel("Exam Score")

# Visualize the relationship between Mental Health Rating and Exam Score 
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='mental_health_rating', y='exam_score', color='green')

# Add title and axis labels
plt.title("Mental Health vs Exam Score", fontsize=14)
plt.xlabel("Mental Health Rating of Students")
plt.ylabel("Exam Score")


# Show the plot
plt.show()
```

```python
# Study Hours vs Exam Score (with trendline)
sns.lmplot(x='study_hours_per_day', y='exam_score', data=df, aspect=1.5)
plt.title('Study Hours vs Exam Score')
plt.xlabel('Study Hours Per Day')
plt.ylabel('Exam Score')
plt.tight_layout()
plt.savefig("study_hours_vs_exam_score.png")
plt.show()
```

```python
#Group Comparison
# Diet Quality vs Exam Score
sns.boxplot(x='diet_quality', y='exam_score', data=df)
plt.title('Diet Quality vs Exam Score')
plt.xlabel('Diet Quality')
plt.ylabel('Exam Score')
plt.tight_layout()
plt.savefig("diet_vs_score.png")
plt.show()

# Gender vs Exam Score
sns.boxplot(x='gender', y='exam_score', data=df)
plt.title('Gender vs Exam Score')
plt.xlabel('Gender')
plt.ylabel('Exam Score')
plt.tight_layout()
plt.savefig("gender_vs_score.png")
plt.show()
```

```python
#Parental Education level vs Exam Score
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='parental_education_level', y='exam_score', palette='Set2')
plt.title('Exam Score by Parental Education Level')
plt.xlabel('Parental Education Level')
plt.ylabel('Exam Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("Parental Education level vs Exam Score")

#Part Time Job vs Exam Score
plt.figure(figsize=(6, 5))
sns.boxplot(data=df, x='part_time_job', y='exam_score', palette='Set1')
plt.title('Exam Score by Part-Time Job Status')
plt.xlabel('Part-Time Job')
plt.ylabel('Exam Score')
plt.tight_layout()
plt.savefig("Part Time Job vs Exam Score")

plt.show()
```

```python
# Create Attendance Buckets
df['attendance_group'] = pd.cut(df['attendance_percentage'], bins=[0, 60, 80, 100], labels=['Low', 'Medium', 'High'])

plt.figure(figsize=(6, 5))
sns.boxplot(data=df, x='attendance_group', y='exam_score', palette='coolwarm')
plt.title('Exam Score by Attendance Level')
plt.xlabel('Attendance Level')
plt.ylabel('Exam Score')
plt.tight_layout()
plt.show()
```
